1st we will try on the openCV 

```python
import cv2
import matplotlib.pyplot as plt

img1 = cv2.imread("images/1.jpg")
image_rgb = cv2.cvtColor(img1, cv2.COLOR_BGR2RGB)
plt.imshow(image_rgb)
```

image_rgb = cv2.cvtColor(img1, cv2.COLOR_BGR2RGB)

we convert here the image to the RGB because the OpenCV read the image as BGR and the matplotlib get the images in the RGB files so we need to convert the images Format First

### Loading Image by Pillow (Python Imaging Library, PIL):

```python
from PIL import Image
import numpy as np

img2 = Image.open("images/2.jpg")
print("Image format:", img2.format)
print("Image mode:", img2.mode)
print("Image size:", img2.size)
img2.show()
```

#### How to plot the image:

```python
img_np =  np.array(img2)

plt.imshow(img_np)
```

### Loading Image by scikit-image:
